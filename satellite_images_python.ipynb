awesome! now we are good to go, run a simple gdal open call and check to see the dimensions of the dataset (note we are only loading 1 band in)

```
raster_ds = gdal.Open("L2A_T30UXB_20170102T111442_TCI_60m.jp2", gdal.GA_ReadOnly)
print raster_ds
image_gdal = raster_ds.GetRasterBand(1).ReadAsArray()
print image_gdal.shape
```

lets take a look at the data, if not already install !pip install matplotlib if install import as useful

```
import matplotlib.pyplot as plt
```

plot the image

```
plt.imshow(image_gdal, cmap = "hot", interpolation='nearest', aspect='auto')
plt.colorbar()
plt.show()
```
